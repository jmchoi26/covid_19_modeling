## The Pendulum

```python
from scipy.integrate import solve_ivp
```

```python
g = 9.81 # gravitational acceleration
l = 1.0 # pendulum length
```

```python
def pendulum_ode(t_span, y0):
    phi, omega = y0
    dphi_dt = omega # change in angle == angular velocity
    domega_dt = -(g/l) * np.sin(phi) # change in angular velocity
    return [dphi_dt, domega_dt]
```

```python
phi0 = np.pi / 4 # angular displacement: angle from vertical
omega0 = 0 # angular velocity: angle change rate
t_span = (0, 10)
solution = solve_ivp(pendulum_ode, t_span, [phi0, omega0], step_size=0.01)
```

```python
plt.figure(figsize=(12, 5))

# Angular displacement over time
plt.subplot(1, 2, 1)
plt.plot(solution.t, solution.y[0], '--')
plt.xlabel('Time (s)')
plt.ylabel('Angle (rad)')
plt.title('Angular Displacement')

# Phase space plot
plt.subplot(1, 2, 2)
plt.plot(solution.y[0], solution.y[1], '--')
plt.xlabel('Angle (rad)')
plt.ylabel('Angular Velocity (rad/s)')
plt.title('Phase Space')

plt.tight_layout()
plt.show()
```
